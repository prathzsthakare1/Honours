## **Collect, Clean, Integrate and Transform Healthcare Data based on specific disease**



**1. Collect**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/Data.csv")
```

```python
print(data)
```

```python
data.head()
```

**2. Clean**

```python
# Convert datetime columns
data['ScheduledDay'] = pd.to_datetime(data['ScheduledDay'])
data['AppointmentDay'] = pd.to_datetime(data['AppointmentDay'])
data.head()
```

```python
# Missing Values
#print(data.isnull())
print('Missing values count:')
print(data.isnull().sum())        #shows count of missing values
print('\nTotal values count:')
print(data.notnull().sum())
```

*No missing values present*

**3. Integrate**

```python
# Convert categorical variables using one-hot encoding
data = pd.get_dummies(data, columns=['Gender', 'Neighbourhood', 'No-show'], drop_first=True)

data.head()
```

**4. Transform**

```python
# Feature engineering (calculate time gap)
data['TimeGap'] = (data['AppointmentDay'] - data['ScheduledDay']).dt.days

print(data['TimeGap'])
```

```python
# Remove irrelevant columns
data.drop(['PatientId', 'AppointmentID', 'ScheduledDay', 'AppointmentDay'], axis=1, inplace=True)

# Print first few rows of the preprocessed data
data.head()
```
