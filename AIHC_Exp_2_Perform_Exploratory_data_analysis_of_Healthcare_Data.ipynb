## **Perform Exploratory data analysis of Healthcare Data**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Univariate Non-graphical**

```python
# Univariate Non-graphical analysis
for column in data.columns:
    if data[column].dtype != 'O':  # Numeric columns
        print(f"Column: {column}")
        print(f"Mean: {data[column].mean()}")
        print(f"Standard Deviation: {data[column].std()}")
        print(f"Minimum: {data[column].min()}")
        print(f"Maximum: {data[column].max()}")
        print(f"25th Percentile (Q1): {data[column].quantile(0.25)}")
        print(f"Median (Q2): {data[column].median()}")
        print(f"75th Percentile (Q3): {data[column].quantile(0.75)}")
        print("\n")
```

**Multivariate Non-Graphical**

```python
#Correlation: find the correlation among the variables
data.corr()
```

**Multivariate Graphical**

```python
# Correlation matrix
#sns.heatmap(data.corr())
plt.figure(figsize=(8, 8))
sns.heatmap(data.corr(), annot=True)
plt.title('Correlation Matrix')
plt.show()
```

**Univariate Graphical**

```python
#Boxplot
#data[['Age']].boxplot()
plt.figure(figsize=(8, 8))
sns.boxplot(data['Age'])
plt.title(f'Box plot of Age')
plt.xlabel('Age')
plt.show()
```

```python
 # Distribution of Age
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
