## **Medical Reviews Analysis from Social Media data**

```python
import requests
from bs4 import BeautifulSoup
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Define the URL of the webpage containing medical reviews
url = 'https://lilavatihospital.com/Testimonials'
```

```python
# Set the user-agent header
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0.0.0 Safari/537.36'}
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
# Send an HTTP GET request to the URL with the custom user-agent header
response = requests.get(url, headers=headers)
```

```python
if response.status_code == 200:
    # Parse the HTML content of the page using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find and extract the elements containing the reviews
    review_elements = soup.find_all('div', class_='testim')

    # Initialize NLTK's sentiment analyzer
    sia = SentimentIntensityAnalyzer()

for review_element in review_elements:
    # Extract the review text from the first <p> element
    review_text = review_element.find('p').text

    # Extract the name and date from the second <p> element
    name_and_date_elements = review_element.find_all('p')
    if len(name_and_date_elements) > 1:
          name_and_date = name_and_date_elements[1].text
    else:
          name_and_date = "Name and Date not available"

    # Tokenize the review text
    tokens = word_tokenize(review_text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

    # Calculate sentiment scores
    sentiment_scores = sia.polarity_scores(review_text)

    print("Review Text:")
    print(review_text)
    print("Name and Date:")
    print(name_and_date)
    print("\nTokens:")
    print(filtered_tokens)
    print("\nSentiment Scores:")
    print(sentiment_scores)
    print("\n---\n")
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```
